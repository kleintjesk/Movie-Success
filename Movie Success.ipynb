```python
import pandas as pd
import numpy as np
```

```python
basics = pd.read_csv('Data/title.basics.tsv.gz', sep='\t', low_memory=False)
```

```python
basics.head()
```

```python
ratings = pd.read_csv('Data/title.ratings.tsv.gz', sep='\t', low_memory=False)
ratings.head()
```

```python
akas = pd.read_csv('Data/title-akas-us-only.csv', low_memory=False)
akas.head()
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
filter_us_titles = basics['tconst'].isin(akas['titleId'])
basics = basics[filter_us_titles]
basics
```

```python
basics = basics.replace({'\\N':np.nan})
```

```python
basics = basics.dropna(subset=['runtimeMinutes', 'genres'])
```

```python
basics = basics.loc[basics['titleType'] =='movie']
basics.head()
```

```python
basics.info()
```

```python
# Convert startYear to a float dtype
basics['startYear'] = basics['startYear'].astype(float)
```

```python
basics.head()
```

```python
#Filter to keep movies with startYears that are >=2000 and <=2022
filter_years = (basics['startYear'] < 2023) & (basics['startYear']>=2000)
```

```python
# remove unwanted years
basics = basics[filter_years]
```

```python
basics.head()
```

```python
#Eliminate movies that include "Documentary" in genre:

filter_documentaries = basics['genres'].str.contains('Documentary')
```

```python
# remove unwanted genre
basics = basics[~filter_documentaries]
```

```python
#Display a final preview of your filtered title basics and save to a csv
basics.head()
```

```python
basics.info()
```

```python
#Load and filter the title ratings file. Load the title ratings data into a dataframe.
filter_basics = ratings['tconst'].isin(basics['tconst'])
ratings = ratings[filter_basics]
```

```python
ratings = ratings.replace({'\\N':np.nan})
```
